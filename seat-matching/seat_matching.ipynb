# Seat Matching Using KMeans Clustering Algorithms given age and income of      the passenger

```python
from numpy import random,array
```

```python
def createClusterData(N,k):
    random.seed(10)
    pointsPerCluster=float(N)/k
    X=[]
    for i in range(k):
        incomeCentroid=random.uniform(20000.0,20000.0)
        ageCentroid=random.uniform(20.0,70.0)
        for j in range(int(pointsPerCluster)):
            X.append([random.normal(incomeCentroid,10000.0),
                     random.normal(ageCentroid,2.0)])
        X=array(X)
        return X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random,float
```

```python
data=createClusterData(1000,5)
```

```python
model=KMeans(n_clusters=5)
```

```python
model=model.fit(scale(data))
```

```python
print(model.labels_)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data[:,0],data[:,1],c=model.labels_.astype(float))
plt.show()
```
